````python
from pathlib import Path

def export_to_md(md_path: str, structure_text: str, *groups: list[str]):
    """
    Gá»™p toÃ n bá»™ code tá»« nhiá»u nhÃ³m file (pages, style_pages, theme_files, v.v.)
    thÃ nh má»™t file Markdown duy nháº¥t.
    """
    md = Path(md_path)
    content = [structure_text.strip(), "\n# ğŸ“œ Code chi tiáº¿t\n"]

    for group in groups:
        for file in group:
            p = Path(file)
            if not p.exists():
                print(f"âš ï¸ Bá» qua {file} (khÃ´ng tá»“n táº¡i)")
                continue

            ext = p.suffix.lower()
            lang = "python" if ext == ".py" else "json" if ext == ".json" else "css"
            code = p.read_text(encoding="utf-8")

            section = (
                f"## `{p.name}`\n"
                f"**Path:** `{p.as_posix()}`\n\n"
                f"```{lang}\n{code}\n```\n"
            )
            content.append(section)
            print(f"âœ… ÄÃ£ thÃªm: {p.name}")

    md.write_text("\n".join(content), encoding="utf-8")
    print(f"\nğŸ“˜ ÄÃ£ táº¡o file Markdown táº¡i: {md.resolve()}")


# --- Gá»i hÃ m ---
export_to_md(
    "prompt.md",
    structure,
    pages,
    style_pages,
    theme_files,
    ui_widgets,
    main_window,
    main,
    pipeline,
    config,
    utils,
)
````
